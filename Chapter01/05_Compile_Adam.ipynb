```python
import tensorflow as tf
import numpy as np
```

```python
from tensorflow import keras
```

```python
EPOCHS = 200
BATCH_SIZE = 128
VERBOSE = 1
NB_CLASSES = 10
N_HIDDEN = 128
VALIDATION_SPLIT = 0.2
DROPOUT = 0.3
RESHAPED = 784
```

```python
mnist = keras.datasets.mnist
```

```python
(X_train, Y_train), (X_test, Y_test) = mnist.load_data()
```

```python
X_train = X_train.reshape(-1, RESHAPED) / 255.
X_test = X_test.reshape(-1, RESHAPED) / 255.
```

```python
Y_train = keras.utils.to_categorical(Y_train, NB_CLASSES)
Y_test = keras.utils.to_categorical(Y_test, NB_CLASSES)
```

```python
model = keras.models.Sequential([
    keras.layers.Dense(N_HIDDEN, input_shape=(RESHAPED, ), activation = 'relu'),
    keras.layers.Dropout(DROPOUT),
    keras.layers.Dense(N_HIDDEN, activation='relu'),
    keras.layers.Dropout(DROPOUT),
    keras.layers.Dense(NB_CLASSES, activation='softmax')
])
```

```python
model.compile(optimizer="Adam", loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train, Y_train, batch_size=BATCH_SIZE, epochs=EPOCHS, validation_split=VALIDATION_SPLIT, verbose = VERBOSE)
```

```python
test_loss, test_acc = model.evaluate(X_test, Y_test)
print('Test accuracy:', test_acc)
```
